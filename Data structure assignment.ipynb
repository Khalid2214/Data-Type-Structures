```python
#Example of a nested dictionary: Store Complex data

nested_dict = {
    'person1': {'name': 'Alice', 'age': 30, 'city': 'New York'},
    'person2': {'name': 'Bob', 'age': 25, 'city': 'Los Angeles'},
    'person3': {'name': 'Charlie', 'age': 35, 'city': 'Chicago'}
}
print(nested_dict)
for i in nested_dict:
  print(nested_dict[i])
#Here, the nested_dict contains dictionaries for each person, with keys like 'name', 'age', and 'city' corresponding to their personal details.
```

Q19 Ans: Dictionaries were considered unordered until Python 3.6. In Python 3.6 and later, dictionaries maintain insertion order, meaning the order of the items is preserved as they were added.

That's how "Unordered" Affect Data Retrieval?
In the context of dictionary operations like data retrieval, the unordered nature of a dictionary (pre-3.6) didn’t really affect how data was accessed by key. The retrieval time for accessing values by key was always O(1) on average, meaning it was very efficient regardless of the order in which the keys were inserted. The unordered property only applied to how the dictionary's elements were iterated or displayed.

Q20 Ans:Difference between list and dict in terms of data retrieval

>>Time Complexity of Data Retrieval:

List:
Accessing an element by index in a list is O(1) on average (constant time).
However, searching for an element by value (not index) takes O(n) time, where n is the number of elements in the list because it has to check each element sequentially.

Dictionary:
Accessing a value by key in a dictionary is O(1) on average (constant time), thanks to the underlying hash table implementation.
Searching by value is O(n), similar to lists, because the dictionary doesn’t directly support searching by values, only by keys.
Example:

>> Order of Data Retrieval:

List:
Lists are ordered collections, meaning the order in which you add elements is preserved.
You retrieve elements in the order of their indexs.

Dictionary:
Dictionaries were unordered until Python 3.6, where insertion order began to be preserved as an implementation detail (guaranteed in Python 3.7+).
However, even with insertion order preserved, dictionaries are primarily designed for fast key-based lookups, not order-based retrieval.


>> Use Cases for Data Retrieval:

List:
Lists are great when:
You need to access elements by position or iterate sequentially over the data.
You want to store and manipulate ordered data.
Searching for elements by their index is needed.


Dictionary:
Dictionaries are ideal when:
You need to retrieve values using a unique key.
You need fast lookups (key-based access) and don’t care about the order.
Data needs to be associated with unique identifiers, such as name-to-age mappings, or product IDs to prices.
